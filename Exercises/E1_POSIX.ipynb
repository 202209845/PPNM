# Exercise "POSIX"
Get yourself a POSIX system and learn its basics.

## Exercise "POSIX" summary

This exercise was completed by setting up a POSIX/Linux environment on Windows using **WSL with Ubuntu**.

The project folder was opened in **Visual Studio Code through WSL**, so all terminal commands and C++ compilation were done inside the Ubuntu/Linux environment. This gives access to a bash terminal, POSIX commands, and the usual GNU development tools.

The basic development tools were installed with `apt`, including

```bash
sudo apt update
sudo apt install build-essential python3 python3-pip python3-venv

In short: **WSL Ubuntu + VS Code + bash terminal + g++/make/Python** solves this exercise.
